```python
import pandas as pd
from scipy.io import arff
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

data, meta = arff.loadarff(r'C:\Users\gzmce\OneDrive - Firat University\Masaüstü\artificalGerekli\chainlink.arff')
df = pd.DataFrame(data)
df
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

# Sayısal verileri al
X = np.column_stack((data['x'], data['y'], data['z']))

# Standartlaştırma
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
data_pd = pd.DataFrame(data)
data_pd["class"] = data_pd["class"].apply(lambda c : c.decode("utf-8"))
data_pd["class"] = data_pd["class"].apply(lambda c : int(c) if c != 'noise' else -1)
data_pd #classları sayısal yaptık.
```

### k means

```python
for col in df.select_dtypes([object]).columns:
    df[col] = df[col].str.decode('utf-8')

X = df.iloc[:, [0, 1, 2]].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.metrics import normalized_mutual_info_score
normalized_mutual_info_score(data_pd["class"],y_kmeans) 
```

```python
import time

start = time.time()

kmeans = KMeans(n_clusters=31, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

end = time.time()
print(f"K-Means çalışma süresi: {end - start:.4f} saniye")
```

### agglomerative

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

for col in df.select_dtypes([object]).columns:
    df[col] = df[col].str.decode('utf-8')

X = df.iloc[:, [0, 1, 2]].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

agg = AgglomerativeClustering(n_clusters=2)
labels_agg = agg.fit_predict(X_scaled)
```

```python
normalized_mutual_info_score(data_pd["class"],labels_agg) 
```

```python
import time

start = time.time()

agg = AgglomerativeClustering(n_clusters=2)
labels_agg = agg.fit_predict(X_scaled)

end = time.time()
print(f"çalışma süresi: {end - start:.4f} saniye")
```

### dbscan

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Kategorik sütunları decode et (gerekliyse)
for col in df.select_dtypes([object]).columns:
    df[col] = df[col].str.decode('utf-8')

# X ve Y özelliklerini al
X = df.iloc[:, [0, 1, 2]].values

# Standardizasyon
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN modelini oluştur
dbscan = DBSCAN(eps=0.2)
labels = dbscan.fit_predict(X_scaled)
```

```python
normalized_mutual_info_score(data_pd["class"],labels) 
```

```python
import time

start = time.time()


dbscan = DBSCAN(eps=0.2)
labels = dbscan.fit_predict(X_scaled)

end = time.time()
print(f"çalışma süresi: {end - start:.4f} saniye")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3D çizim için
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# X_scaled[:,0], X_scaled[:,1], X_scaled[:,2] 3 özellik
scatter = ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2], c=labels, cmap='viridis')

ax.set_title('DBSCAN Kümeleme')
ax.set_xlabel('Özellik 1 (x)')
ax.set_ylabel('Özellik 2 (y)')
ax.set_zlabel('Özellik 3 (z)')

fig.colorbar(scatter, ax=ax, label='Küme Etiketi')

plt.show()
```
